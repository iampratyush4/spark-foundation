```python
#importing python libraries 

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('darkgrid') # to support warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# to get top 5 results of the dataframe
gt = pd.read_csv("globalterrorismdb_0718dist.csv",encoding='latin1')
gt.head()
```

```python
#to get the values of all the columns
gt.columns.values
```

```python
# to get information about the dataframe
gt.info()
```

```python
# to view some basic statical details like percentile, mean, std etc of a dataframe
gt.describe()
```

```python
#to remove the column with NaN values
gt.dropna(how='all',axis=1)
gt.head()
```

```python
#to rename the existing columns
gt.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','gname':'Group','country_txt':'countrylist','region_txt':'Region',
          'provstate':'State','city':'City','latitude':'latitude','longitude':'longitude','summary':'summary',
          'attacktype1_txt':'Attacktype','targtype1_txt':'aim','weaptype1_txt':'Weapon','nkill':'kill',
          'nwound':'Wound'},inplace=True)
```

```python
gt.head()
```

```python
#filtering the important columns
gt = gt[['Year','Month','Day','countrylist','State','Region','City','latitude','longitude','Attacktype','kill','Wound','target1',
         'summary','Group','Weapon','target1','motive']]
```

```python
# to find the total number of null value in the column
gt.isnull().sum()
```

```python
#to plot total number of attacks vs year
year = gt['Year'].unique()
years_count = gt['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (15,10))
sns.barplot(x = year, 
            y = years_count, 
            palette = "tab10")
plt.xticks(rotation=90)
plt.xlabel('Year',fontsize=20)
plt.ylabel('Number of Attacks',fontsize=20)
plt.title('Total number of attacks in years',fontsize=40)
plt.show()
```

## Maximum number of attacks in year 2014

```python
gt1 = gt[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(16,10))
gt1.plot(kind='bar',alpha=1,ax=ax4)
plt.xticks(rotation=90)
plt.title("People Died in Attack",fontsize=25)
plt.ylabel('Number of people killed',fontsize=20)
plt.xlabel('Year',fontsize=20)
```

```python
# to plot total number of people wounded vs year
gt1 = gt[['Year','Wound']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(16,10))
gt1.plot(kind='bar',alpha=1,ax=ax4)
plt.xticks(rotation=90)
plt.title("People wounded in Attack",fontsize=40)
plt.ylabel('Number of people killed',fontsize=20)
plt.xlabel('Year',fontsize=20)
```

```python
# to count the top 5 country which faced maximum attacks
attack = gt.countrylist.value_counts()[:5]
attack
```

```python
# to count the top 5 city which faced maximum attacks
attack = gt.City.value_counts()[:5]
attack
```

```python
# to plot total number of attacks bs city
gt['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).drop('Unknown').head(15).plot(kind='bar',figsize=(30,15),color='red')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 15 most effected city",fontsize=20)
plt.show()
```

```python
# to count the regions which faced maximun attacks
attack = gt.Region.value_counts()[:5]
attack
```

```python
# to plot total number of attacks vs region
pd.crosstab(gt.Year, gt.Region).plot(kind='area',figsize=(30,15))
plt.title('Terrorist Activities by Region in each year')
plt.ylabel('Region of Attacks')
plt.show()
```

```python
# to plot total number of attacks vs attack_type
gt['Attacktype'].value_counts().to_frame().sort_values('Attacktype',axis=0,ascending=False).drop('Unknown').head(15).plot(kind='bar',figsize=(20,15),color='red')
plt.xticks(rotation = 90)
plt.xlabel("Attack_type",fontsize=20)
plt.ylabel("Number of attack",fontsize=20)
plt.title("Attack_type Count",fontsize=30)
plt.show()
```

```python

# to plot total number of attack vs groups
gt['Group'].value_counts().to_frame().drop('Unknown')[:20].plot(kind='bar',figsize=(20,10),color='red')
plt.title("Top 20 terrorist group attack",fontsize=40)
plt.xlabel("terrorist group",fontsize=15)
plt.ylabel("Attack nummber",fontsize=15)
plt.show()
```
